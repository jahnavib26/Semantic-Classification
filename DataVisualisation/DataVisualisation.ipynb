```python
import json
import matplotlib.pyplot as plt

# Load data from JSON file
file = open("data_file_2L.json", 'r')
json_data = json.load(file)
file.close()

```

# ***Data Before Imputation***

**Total Null Values**


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Excel sheet into a pandas dataframe
df = pd.read_excel("Datset200k.xlsx")
# Calculate the total number of null values in each column
null_counts = df.isnull().sum()

# Create a bar graph of the null value counts
plt.bar(null_counts.index, null_counts.values)

# Add labels and title to the graph
plt.xticks(rotation=90)
plt.xlabel('Column')
plt.ylabel('Number of Null Values')
plt.title('Total Number of Null Values ')

# Display the graph
plt.show()
```

**Pie-chart of occurance of different Genders and Name titles**

```python
import matplotlib.pyplot as plt

# Specify the columns you want to plot
columns_to_plot = ['NAME_TITLE', 'GENDER']

# Loop through the specified columns and count the occurrences of each value
for column in columns_to_plot:
    counts = df[column].value_counts()

    # Create a pie chart of the counts
    plt.pie(counts.values, labels=counts.index, autopct='%1.1f%%')

    # Add title to the graph
    plt.title(f'Occurrence of Values in Column {column}')

    # Display the graph
    plt.show()
```

**Relationship between the occurance of Name_title and Gender**

```python
# Convert float columns to strings
df['NAME_TITLE'] = df['NAME_TITLE'].astype(str)
df['GENDER'] = df['GENDER'].astype(str)

# Specify the columns to plot 
columns_to_plot = ['NAME_TITLE', 'GENDER',] 

# Create a scatter plot 
plt.scatter(df[columns_to_plot[0]], df[columns_to_plot[1]])

# Set axis labels 
plt.xlabel(columns_to_plot[0]) 
plt.ylabel(columns_to_plot[1]) 


# Show the plot 
plt.show()
```

**Bar graph of total number of people from each state**

```python
# Specify the columns you want to plot
columns_to_plot = ['STATE']

# Loop through the specified columns and count the occurrences of each value
for column in columns_to_plot:
    counts = df[column].value_counts()

    # Create a bar graph of the counts with wider bars
    plt.bar(counts.index, counts.values, width=0.5)

    # Add labels and title to the graph
    plt.xticks(rotation=90)
    plt.xlabel('States')
    plt.ylabel('Count')
    plt.title(f'Occurrence of Values in Column {column}')
    plt.tight_layout()

    # Display the graph
    plt.show()
```

## ***Data after Imputation*** 

**Occurance of each datapoint after Imputation**

```python
merged_dict = {}
for dictionary in json_data:
    for key, value in dictionary.items():
        if key not in merged_dict:
            merged_dict[key] = []
        merged_dict[key].append(value)

# Create a bar chart of the merged data
plt.figure(figsize=(10,5))
plt.bar(list(merged_dict.keys()), [len(values) for values in merged_dict.values()])
plt.xticks(rotation=90)
plt.xlabel('Fields')
plt.ylabel('Number of Values')
plt.title('Occurance of Data points (AFTER IMPUTATION)')
plt.show()
```

**Maximum lenght of characters in each column**

```python
max_lengths = {}

# Loop over the data and find the maximum length for each column
for dictionary in json_data:
    for key, value in dictionary.items():
        if key not in max_lengths:
            max_lengths[key] = 0
        max_lengths[key] = max(max_lengths[key], len(value))

# Plot a bar graph
plt.bar(max_lengths.keys(), max_lengths.values())
plt.xlabel("Key Name")
plt.ylabel("Maximum Length")
plt.title("Maximum Length in each column")
plt.xticks(rotation=45, ha='right') # Rotate x-axis tick labels by 45 degrees and align them to the right
plt.tight_layout() # To fix layout issues, if any
plt.show()
```
